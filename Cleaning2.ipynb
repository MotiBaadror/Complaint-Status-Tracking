```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
data=pd.read_csv('Traning1.csv',index_col='Complaint-ID')
data.head()
```

```python
compl_reas=data['Complaint-reason'].unique()
compl_reas.shape
a=np.arange(1,151)
data['Complaint-reason']=data['Complaint-reason'].map(dict(zip(compl_reas,a)))
data.head()
```

```python
data['Company-response']=data['Company-response'].replace(1,4)
# data['Company-response'].value_counts()
data.head()
```
